```python
###################################project2 transfer curency from country to other ####################################3
from scrapy import Selector
import requests
import re 
from datetime import datetime 
base_url="https://www.xe.com/currencyconverter/convert/?"
From = input("convetred from : ")
to =input("to : ")
much = float(input("how much you want to convert from {0} to {1} : ".format(From , to)))
url = base_url+"Amount="+str(much)+"&"+f"From={From}"+"&"+f"To={to}"
response = requests.get(url)
if (response.ok is False):
    print("\nError {0}".foramt(response.status_code))
    print(response.json()["error"])
else :
    html = Selector(response)
    res=html.xpath('//*[@id="__next"]/div[2]/div[2]/section/div[2]/div/main/form/div[2]/div[1]/p[2]')[0].extract()
    result =re.findall(">(.+)<span" , res)[0]
    print(datetime.now())
    print("the quantinity {0} from {1} to {2} is equal ={3} ".format(much , From , to , result))
```
